```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
np.set_printoptions(suppress=True)

% matplotlib inline

```

```python
# Original Data Loading

data=pd.read_csv('finaldata.csv',index_col=False)
data.head()
```

```python
# Define a method to clean the original data set and develop a linear regression model

def get_regression(target,Stratification):
    dataframe = data[(data['Question']==target)&(data['StratificationCategory1']==Stratification)]
    dataframe['Total Sq. Mi.'] = dataframe['Total Sq. Mi.'].str.replace(',','').astype('int')
    if Stratification == 'Overall':
        pass
    else:
        dataframe = dataframe.join(pd.get_dummies(dataframe.Stratification1))
    
    dataframe = dataframe.drop(dataframe.columns[[0,1,2,11,13,14,17]],axis=1)
    dataframe = dataframe.dropna(axis=0)
    X = dataframe.drop(['DataValue'],axis=1) 
    y = dataframe[['DataValue']]
        
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
    linreg = LinearRegression().fit(X_train, y_train)
    print('Score for linear regression on'+' '+target)
    print(linreg.score(X_train,y_train))
    print(linreg.score(X_test,y_test))
    print('\n')
```

```python
# Run the model on selected diseases and focus on the effect of race/ethnicity

target_list = ["Asthma mortality rate", 
               "Mortality from diseases of the heart", 
               "Invasive cancer (all sites combined), incidence"]
for tar in target_list:
    get_regression(target=tar, Stratification='Race/Ethnicity')
    
```

When we focus on racial variable, the accuracy of linear regression model is about 70% - 80% on the test set

```python
# For comparison, what will happen if we don't take race/ethnicity variable apart 

for tar in target_list:
    get_regression(target=tar, Stratification='Overall')
    
```

However, if we don't take racial variable apart from the original data, the accuracy of linear regression is extramely low.
