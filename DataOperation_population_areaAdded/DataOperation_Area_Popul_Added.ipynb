# DataOperation - Addind Population and Area columns


```python
import pandas as pd
```

## Understanding various aspects of population and existing dataset

```python
finaldata = pd.read_csv('finaldatatouse.csv')
popul_area = pd.read_csv('popul.csv')
```

```python
finaldata.head()
```

```python
df = finaldata.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
df[60:80]
```

```python
popul_area.head()
```

```python
popul_area = popul_area.drop(columns=['NAME','Total Sq. Mi.','Total Sq. Km.'])
```

```python
popul_area.head()
```

```python
popul_area = popul_area[:-1]
```

## Using melt function to pivot required columns

```python
popul_area = pd.melt(popul_area, id_vars=['State'], 
        var_name='Year', value_name='Population')
```

```python
popul_area.tail()
```

```python
popul_area = popul_area.rename(columns = {'State':'STATE'})
popul_area.head()
```

```python
popul_area_mod = popul_area
```

```python
popul_area_mod.head()
```

```python
popul_area_mod['Year'] = popul_area_mod['Year'].str.replace(r'\D', '').astype(int)
```

```python
popul_area_mod[50:60]
```

```python
popul_area_mod.head()
```

```python
df.head()
```

## Doing merge after checking for correctness of the datasets

```python
new_data = pd.merge(
                     df,
                     popul_area_mod,
                     how  = 'left', on = ['STATE','Year'])
new_data.tail()
```
