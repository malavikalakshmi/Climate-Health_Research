```python
import pandas as pd
final = pd.read_csv('US_Chronic_Disease_Indicators.csv')
```

```python
final.head()
```

```python
final_modified_datasource = final.loc[(final['DataSource'] == 'NVSS') | (final['DataSource'] == 'Statewide central cancer registries')]
final_modified_datasource.info()
```

```python
final_modified_datasource = final_modified_datasource.loc[(final_modified_datasource['YearStart'] == 2012)|(final_modified_datasource['YearStart'] == 2013) |(final_modified_datasource['YearStart'] == 2014)|(final_modified_datasource['YearStart'] == 2015)|(final_modified_datasource['YearStart'] == 2016)|(final_modified_datasource['YearStart'] == 2017)|(final_modified_datasource['YearStart'] == 2018)]
final_modified_datasource.info()
```

```python
final_modified_datasource = final_modified_datasource.loc[(final_modified_datasource['Topic'] == 'Overarching Conditions') |(final_modified_datasource['Topic'] == 'Asthma') |(final_modified_datasource['Topic'] == 'Cardiovascular Disease') | (final_modified_datasource['Topic'] == 'Cancer') | (final_modified_datasource['Topic'] == 'Chronic Obstructive Pulmonary Disease')]
final_modified_datasource.info()
```

```python
final_modified_datasource = final_modified_datasource.loc[(final_modified_datasource['Question'] == 'Asthma mortality rate') | (final_modified_datasource['Question'] == 'Cancer of the lung and bronchus, incidence') | (final_modified_datasource['Question'] == 'Invasive cancer (all sites combined), incidence') | (final_modified_datasource['Question'] == 'Mortality from coronary heart disease') | (final_modified_datasource['Question'] == 'Mortality from diseases of the heart') | (final_modified_datasource['Question'] == 'Mortality from heart failure') | (final_modified_datasource['Question'] == 'Mortality from total cardiovascular diseases') | (final_modified_datasource['Question'] == 'Mortality with chronic obstructive pulmonary disease as underlying cause among adults aged >= 45 years') | (final_modified_datasource['Question'] == 'Mortality with chronic obstructive pulmonary disease as underlying or contributing cause among adults aged >= 45 years')| (final_modified_datasource['Question'] == 'Premature mortality among adults aged 45-64 years')]  
```

```python
final_modified_datasource.info()
```

```python
final_modified_datasource = final_modified_datasource.loc[(final_modified_datasource['DataValueType'] == 'Crude Rate') | (final_modified_datasource['DataValueType'] == 'Average Annual Crude Rate')]
```

```python
final_modified_datasource.info()
```

```python
final_modified_datasource.to_csv('new_final_modified_datasource.csv')
```
